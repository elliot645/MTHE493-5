```python
db = pd.read_csv('C:/Users/ellio/python/MTHE493-5/data/us_county_population_estimates_by_age_2020_to_2023.csv', encoding='latin1')

db.head(10)
```

```python
data = {}
for index, row in db.iterrows():
    year_dict = data.setdefault(row['YEAR'], {})
    state_dict = year_dict.setdefault(row['STNAME'], {})
    state_dict[row['CTYNAME']] = row['POPESTIMATE']

    year_dict[row['STNAME']] = state_dict
    data[row['YEAR']] = year_dict
```
